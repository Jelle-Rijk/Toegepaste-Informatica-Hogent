# Basisconcepten in data science

## Variabelen en meetniveaus

Variabele = een waarde van een attribuut van het bestudeerde onderwerp

Worden onderverdeeld in meetniveaus:

- Kwalitatieve / Categorische variabelen: Beperkt aantal mogelijke waarden, niet altijd numeriek
  - Nominaal: Waarden hebben geen logische volgorde (vb. kleuren, landen)
  - Ordinaal: Waarden hebben wel een logische volgorde (vb. beoordeling op 5 sterren, hoogst behaalde diploma)
- Kwantitatieve / Continue variabelen: Altijd numeriek en met een meeteenheid. Je verwacht veel verschillende waarden.
  - Intervalschaal: Geen vast of duidelijk nulpunt, men is op zoek naar verschillen tussen waarden (vb. temperatuur in celsius, datum, tijdstip).
  - Ratioschaal: Duidelijk vast nulpunt. Hierdoor kan je werken met verhoudingen (vb. afstanden, leeftijden, energie, temperatuur in Kelvin).

Computers zijn niet goed in bepalen van meetniveaus. Meetniveaus zijn belangrijk om juiste visualisaties en statistische methodes te kiezen.

*Bij examenvragen: noteer nominaal, ordinaal, interval of ratio (kwalitatief/kwantitatief is niet nodig)*

## Populaties, steekproeven, observaties

- **Populatie** = de hele set individuen of voorwerpen die bestudeerd worden.
- **Steekproef** / sample = deel van de populatie die gebruikt wordt voor analyse.
- **Observatie** = één individu of voorwerp uit de steekproef.

Een goede steekproef is:
- Aselect / Randomly selected: Elk individu in de populatie heeft even veel kans om gekozen te worden.
- Groot genoeg
- Representatief voor de karakteristieken van de hele populatie (vb. ratio van genders, leeftijden, etc.).

Steekproefkader / Sampling frame = lijst met alle individuen van de populatie.

### Steekproeffouten / Sampling errors

Fouten in de manier waarop je een steekproef samenstelt. Omdat je met een steekproef werkt, zitten er onvermijdelijk fouten in de data. Daarom is het belangrijk de fouten in categorieën in te delen.

- Willekeurige steekproeffouten: Kunnen niet vermeden worden, zijn een gevolg van een deel van de populatie te gebruiken.
- Systematische steekproeffouten: De manier van steekproeven uitvoeren zorgt voor een bias in de keuze van de samples (vb. online survey -> enkel mensen met internetverbinding)

Niet-steekproeffouten zijn fouten die gebeuren na het nemen van de sample (dus in de meting of analyse) -> willekeurig (vb. fout ingevuld) / systematisch (vb. meetsysteem verkeerd ingesteld / mensen verbloemen hun rookgedrag).

### Gestratificeerde steekproeven

Om een representatieve steekproef te bereiken worden soms strata gemaakt. Dit zijn deelpopulaties die aan bepaalde criteria voldoen (vb. bepaalde leeftijd).


## Univariate vs. multivariate analyse

Univariate analyse = focus op één variabele <br>
Multivariate analyse = focus op meerdere variabelen

Als één variabele informatie geeft over een andere variabele is er een associatie/relatie. De variabele die informatie geeft is de **onafhankelijke** variabele, de andere variabele is de **afhankelijke** variabele.

Belangrijk om in het achterhoofd te houden: correlatie is geen causatie.
