# <a id='toc2_'></a>[Lies, damn lies and statistics](#toc0_)

Statistieken kunnen misleidend zijn als je een verkeerde plotmethode kiest:

- inconsistente ijking
- data op y-as start niet van nul (vertekend beeld van hoe groot veranderingen zijn)
- verschillende weergave van data (vb. 21% voor / 79% tegen)

# <a id='toc3_'></a>[Basisconcepten in data science](#toc0_)

## <a id='toc3_1_'></a>[Variabelen en meetniveaus](#toc0_)

Variabele = een waarde van een attribuut van het bestudeerde onderwerp

Worden onderverdeeld in meetniveaus:

- Kwalitatieve / Categorische variabelen: Beperkt aantal mogelijke waarden, niet altijd numeriek
  - Nominaal: Waarden hebben geen logische volgorde (vb. kleuren, landen)
  - Ordinaal: Waarden hebben wel een logische volgorde (vb. beoordeling op 5 sterren, hoogst behaalde diploma)
- Kwantitatieve / Continue variabelen: Altijd numeriek en met een meeteenheid. Je verwacht veel verschillende waarden.
  - Intervalschaal: Geen vast of duidelijk nulpunt, men is op zoek naar verschillen tussen waarden (vb. temperatuur in celsius, datum, tijdstip).
  - Ratioschaal: Duidelijk vast nulpunt. Hierdoor kan je werken met verhoudingen (vb. afstanden, leeftijden, energie, temperatuur in Kelvin).
